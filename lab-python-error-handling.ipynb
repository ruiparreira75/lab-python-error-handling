````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Pleaase enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity in numbers.")
    return inventory

inventory = initialize_inventory(products)



    
````

3. Modify the `get_customer_orders` function to include error handling.
   - If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the number of orders.
   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

````python
def get_customer_orders(products, inventory):
    avaiable_products = [p for p in products if inventory.get(p, 0) > 0]
    
    while True:
        try:
            num_orders = int(input("How many different products would you like to order? ")) 
            if num_orders <= 0:
                print("Number of products must be greater than zero.")
                continue
            if num_orders > len(avaiable_products):
                print(f"You can only order up to {len(avaiable_products)} different products (based on the stock).")
                continue
            break
           
        except ValueError:
            print("Please enter a valid number.")
    
    customer_orders = set()
    
    for i in range(num_orders):
        while True:
            try:
                product_name = input(f"Enter product {i + 1} name: ").strip().lower() 
                if not product_name:
                    raise ValueError("Product name cannot be empty.")
                if product_name not in products:
                    raise ValueError(f"'{product_name}' is not in list of products.")
                if inventory.get(product_name, 0) <= 0:
                    raise ValueError(f"'{product_name}' is currently out of stock.")
                if product_name in customer_orders:
                    print(f"'{product_name}' is already in your orders. Please enter a different product.")
                    continue

                customer_orders.add(product_name)
                print(f"Added '{product_name}' to your orders.")
                break
            except ValueError as e:
                print(e)
                print("Please try again.")

    return customer_orders
                
customer_orders = get_customer_orders(products, inventory)
print("Final Customer Orders:", customer_orders)
````

````python
def update_inventory(inventory, customer_orders):
    updated_inventory = {
        product: (inventory[product] - 1 if product in customer_orders and inventory[product] > 0 else inventory[product]) for product in inventory}
    return updated_inventory

updated_inventory = update_inventory(inventory, customer_orders)
````

````python
def print_updated_inventory (inventory):
    formatted_inventory = ", ".join(f"{product}: {quantity}" for product, quantity in inventory.items())
    print("the updated inventory:", formatted_inventory)

print_updated_inventory(updated_inventory)
````

2. Modify the `calculate_total_price` function to include error handling.
   - If the user enters an invalid price (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the price for that product.
   - Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

````python
def get_order_prices(customer_orders):
    prices = {}
    total = 0.0
    
    for product in customer_orders:
        valid_input = False
        while not valid_input:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price >= 0:
                    prices[product] = price
                    total += price
                    valid_input = True
                else:
                    print("Price cannot be negative. Please enter a valid price.")
            except ValueError:
                    print("Invalid input. Please enter a valid price.")                                    
    return prices, total

prices, total = get_order_prices(customer_orders)
print("Order prices:", prices)
print("Total price:", round(total, 2))
````

````python
def update_inventory(inventory, customer_orders):
    for product in customer_orders:
        if product in inventory and inventory[product] > 0:
            inventory[product] -= 1
    
    updated_inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    return updated_inventory

updated_inventory = update_inventory(inventory, customer_orders)
print("The updated inventory:", updated_inventory)
````

````python
def calculate_order_statistics(customer_orders, products):
    total_products_order = len(customer_orders)
    percentage = len(set(customer_orders)) / len(products) * 100
    return total_products_order, percentage
    
update_inventory(inventory, customer_orders)
total, percentage = calculate_order_statistics(customer_orders, products)
````

````python
def print_order_statistics(order_statistics):   
    total, percentage = order_statistics
    print("The total products ordered: ", total)
    print("The percentage of unique products ordered are: ", round(percentage, 2),"%")
    
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
````
